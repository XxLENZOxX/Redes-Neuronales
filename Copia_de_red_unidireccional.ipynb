#**Paso 1: Importar las bibliotecas necesarias** descarga (1).jfif

```python
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
```

#**Paso 2: Cargar el conjunto de datos MNIST** _-_.jfif

```python
mnist = tf.keras.datasets.mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

#**Normalizar los valores de pixeles entre 0 y 1** https___twitter_com_sh46511596_status_846845426753941504.jfif

```python
train_images = train_images / 255.0
test_images = test_images / 255.0
```

#**Paso 3: Construir el modelo unidireccional (FeedForward)** 17878161-bab4-4f93-84a9-ae05d1f69384.jfif

```python
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),  # Aplanar las imgenes 28x28 en vectores de 784 elementos
    layers.Dense(128, activation='relu'),  # Capa densa con 128 neuronas y funcin de activacin ReLU
    layers.Dense(10, activation='softmax') # Capa de salida con 10 neuronas (una por dgito), usando softmax
])
```

#**Paso 4: Compilar el modelo** 44046d50-6788-410b-b0f0-b119d59001bf.jfif

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

#**Paso 5: Entrenar el modelo** descarga (3).jfif

```python
history = model.fit(train_images, train_labels, epochs=5, validation_data=(test_images, test_labels))
```

#**Paso 6: Evaluar el modelo**    .jfif

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)
print(f'\nPrecisin en el conjunto de prueba: {test_acc}')
```

```python
plt.plot(history.history['accuracy'], label='Precisin en el entrenamiento')
plt.plot(history.history['val_accuracy'], label='Precisin en la validacin')
plt.xlabel('pocas')
plt.ylabel('Precisin')
plt.legend(loc='lower right')
plt.show()
```
